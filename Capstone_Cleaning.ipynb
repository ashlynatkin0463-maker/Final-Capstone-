```python
from ucimlrepo import fetch_ucirepo
import pandas as pd

```

```python
# Fetch dataset
apartment_for_rent_classified = fetch_ucirepo(id=555)

# Features and target
X = apartment_for_rent_classified.data.features
y = apartment_for_rent_classified.data.targets

# Combine into one DataFrame
df = pd.concat([X, y], axis=1)

# View result
print(df.head())
print(df.shape)
```

```python
import pandas as pd
import numpy as np

df_clean = df.copy()

# Numeric columns
df_clean['bathrooms'] = pd.to_numeric(df_clean['bathrooms'], errors='coerce')
df_clean['bedrooms'] = pd.to_numeric(df_clean['bedrooms'], errors='coerce')
df_clean['square_feet'] = pd.to_numeric(df_clean['square_feet'], errors='coerce')

# Boolean-like columns
df_clean['fee'] = df_clean['fee'].map({
    'Yes': True,
    'No': False,
    'yes': True,
    'no': False
})

# Unix timestamp -> datetime (if that's what time contains)
df_clean['time'] = pd.to_datetime(
    df_clean['time'],
    unit='s',
    errors='coerce'
)
```
